## 1. Importing CSV data into Python-Exploring Options

## 2. How to Generate Data Frame Summary Statistic

```python
# How to make dataframe using excel sheet
import pandas as pd
excel=pd.read_excel('cereal.xls',0)
excel.head(2)
```

```python
# It will show data of those columns which have only numeric data
excel.describe()
```

```python
excel.describe(include='all')
```

```python
# It contains the columns which are having only charcter values.
excel.describe(include=['object'])
```

```python
# It your excel contains some other data type values and we want that specific values, so we can provide data type also.
excel.describe(include=['object','float'])
```

```python
# isme mene fibre column ki aisi values show ki jo 5 se badi ho
# in this way we can get specific values related to a column
excel[excel['fiber']>5].describe()
```

```python
excel[excel['fiber']>5]
```

```python
# multiple conditions can be applied in the following manner
excel[(excel['fiber']>5) & (excel.carbo>10.0)]
```

```python
excel[(excel['fiber']>5) & (excel.carbo>10.0)].describe()
```

## 3. How to sort values in a dataframe

```python
# we are sorting values of sodium column in ascending order
excel.sodium.sort_values()
```

```python
excel.sodium.sort_values(ascending=False).head()
```

```python
# sorting columns that has space
excel['sodium'].sort_values().head()
```

```python
# sorting the entire dataframe in ascending order and here we are providing impact on column name 
excel.sort_values('name').head(10)
```

```python
# sorting the entire dataframe in descending order ,
# and values of all the columns are arranged according to the values given in protein column
excel.sort_values('protein',ascending=False).head(10)
```

```python
# sorting multiple columns
excel.sort_values(['calories','sodium']).head(10)
```

```python
# descending order
excel.sort_values(['type','calories'],ascending=False).head(10)
```

## 4. Applying filters to Dataframe

```python
# if we want values of 'calories' column greater than 100
excel[excel.calories>100].head(10)
```

```python
excel[excel['type']!= 'C']
```

```python
excel[(excel['fat'] !=0) & (excel['calories']>=100)].head(10)
```

```python
excel[(excel['type'] !='C') | (excel['calories']>=100)].head(10)
```

## 5. How to delete columns and rows from dataframe

```python
# inplace=True means this changes are permemnately made into the datframe
# axis=1 means column, if we want to delete column then put axis=1, otherwise put axis=0
excel.drop(['mfr'],axis=1,inplace=True)
```

```python
excel.head(10)
```

```python
# To remove two diferent columns
excel.drop(['type','rating'],axis=1,inplace=True)
excel.head(10)
```

```python
# to delete rows ,write axis=0
excel.drop([0],axis=0,inplace=True)
```
