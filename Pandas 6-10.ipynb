## 6. How to manipulate pandas DataFrame

```python
import pandas as pd
from pandas import DataFrame
df=pd.read_csv('Iris.csv')
```

```python
df.head(20)
```

```python
# let's review all the columns present in a dataset
df.columns
```

```python
# let's create a subset of a dataset
df_subset=DataFrame(df,columns=['SepalLengthCm','SepalWidthCm'])
df_subset.head()
```

```python
# we can also add some extra column while creating a subset
df_subset=DataFrame(df,columns=['SepalLengthCm','SepalWidthCm','new'])
df_subset.head()
```

```python
# how to get individual column
df['SepalLengthCm']
```

```python
# agar column name ki spelling me space hota he toh df['column_name'] use krte he
# aur agar nhi hota toh df.column_name, for ex-
df.Species
```

```python
# Deleting column
del df_subset['new']
df_subset.head()
```

## 9. How to import excel data in Python

```python
import pandas as pd
```

```python
xl_file=pd.ExcelFile('Sample - Superstore Sales.xls')
```

```python
xl_file.head()
```

```python
# here we are creating an object 'orders', and we want only 'Orders' sheet.
# So we are parsing 'Orders' sheet into 'orders' object
orders=xl_file.parse('Orders')
orders.head()
```

```python
# another way to read data from excel file
returns=pd.read_excel('Sample - Superstore Sales.xls')
returns.head()
```

## 10. Answering business questions with pandas/ Exploring data analysis

```python
import pandas as pd
```

```python
df1=pd.read_excel('Sample - Superstore Sales.xls')
df1.head(10)
```

```python
df1.groupby('Customer Segment').Sales.sum()
```

```python
df1.groupby('Customer Segment').Profit.sum()
```

```python
#how many times a product appears 
df1['Product Name'].value_counts()
```

```python
# how many unique products you have
df1['Product Name'].nunique()
```

```python
# common opeartion in analysis
# identify the top 5 product name or sales
df1.groupby('Product Name').Profit.sum().nlargest(5)
```

```python
# 5 lowest product 
df1.groupby('Product Name').Profit.sum().nsmallest(5)
```
